```python
# Logical Function

x = np.array([10, 20, 30, 40, 100])
y = np.array([50, 70, 80, 90, 100])
ans = x > y
print(ans)

result= x == y
print(result)



print(np.all(ans))
print(np.any(result))

```

```python
# Sorting (Inplace) :
x = np.array([10, 8, 18, 16, 100])
z = x.copy()
z.sort()
print(x)
print(z)

# Copy Sorting :
print(x)
sorted = np.sort(x)
print(sorted)
print("___________________________________")
print()

# 2D

mat = np.array([
    [10, 3, 5],
    [8, 4, 9]
    ])

# horizontal axis = 1
horizon_sort = np.sort(mat, axis=1)
print(horizon_sort)

print("--------------")


# vertical axis = 0
vertical_sort = np.sort(mat, axis=0)
print(vertical_sort)
```

```python
# Searching :
# np.where(condition, x, y) if True : x, otherwise replace with y

print(x)
index = np.where(x==16)
print(index)
print()

index = np.where(x>8)
print(index)
print()

index = np.where(x>8, x, -1)
print(index)
print()


print(mat)
index = np.where(mat>8)
print(index)
print()

index = np.where(mat>8, mat, -1)
print(index)
```

```python
# Counting :

a = np.random.randint(1, 100, (100))
print(a)

valGret60 = np.count_nonzero(a==50)
print(valGret60)

unique_val, count = np.unique(a, return_counts=True)
print(unique_val)
print(count)
```

# Statistical Methods

```python
data = np.genfromtxt('/content/sample_data/student_scores.csv', delimiter=',', skip_header=True)

math_marks = data[::, :1]
print(math_marks.T)

max_math_marks = np.max(math_marks)
print(max_math_marks)

min_math_marks = np.min(math_marks)
print(min_math_marks)

average_math_marks = np.mean(math_marks)
print(average_math_marks)

median_math_marks = np.median(math_marks)
print(median_math_marks)

std_math_marks = np.std(math_marks)
print(std_math_marks)
```

```python
study_hour = np.array([2, 4, 5, 7, 8])
exam_hour = np.array([65, 75, 78, 88, 92])

data = np.array([study_hour, exam_hour])

correlation = np.corrcoef(data)
print("Correlation Matrix:\n", correlation)
```

```python
# Linear Algebra :

A = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

B = np.array([
    [7, 8],
    [9, 10],
    [11, 12]
])

dot_product = np.dot(A, B)
print(dot_product)

print(np.trace(B)) # Diagonal Sum.


sq_mat = np.array([
    [1, 24, 3],
    [4, 51, 6],
    [7, 82, 9]
])

det_of_sq_mat = np.linalg.det(sq_mat)
print(det_of_sq_mat)

rank_of_sq_mat = np.linalg.matrix_rank(sq_mat)
print(rank_of_sq_mat)
```
